# Von Mises Stress Quicklook
Load the voxel-averaged von Mises field generated by the solid benchmark
and inspect a mid-volume slice.

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

bundle_root = Path('..') / '20_bench_solid'
volume_path = bundle_root / 'von_mises_t000400.npz'
data = np.load(volume_path)
volume = data['volume']
mid = volume.shape[2] // 2
plt.figure(figsize=(6, 6))
plt.imshow(volume[:, :, mid], origin='lower', cmap='viridis')
plt.colorbar(label='von Mises (Pa)')
plt.title(f'Mid-slice (index {mid}) of von Mises volume')
plt.tight_layout()
plt.show()
```
